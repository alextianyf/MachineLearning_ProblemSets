## 1. Gather and Clean Data
Movie Budget and Gross data is provided by [The Numbers](https://www.the-numbers.com/movie/budgets/all)<br>
the unfiltered dirty data is provided as `cost_revenue_dity.csv` in this folder. However, you are required to delete the disputed figures manually(deleting 0$ worldwide Gross rows, Domestic Gross coloumn, Rank coloumn, Release Data coloumn, Movie Title coloumn, '$' sign, and rename `Production_Budget_USD` and `Worldwide_Gross_USD`).

```python
#If you don't have pandas install, please execute 'pip install pandas' in CMD to install the package
import pandas as pd
from pandas import DataFrame
#If you don't have matplotlib install, please execute 'python -m pip install -U matplotlib' in CMD to install the package
import matplotlib.pyplot as plt
#install with command 'pip install scikit-learn'
from sklearn.linear_model import LinearRegression
```

```python

data = pd.read_csv('cost_revenue_cleaned.csv')
data
```

```python
data.describe()
```

```python
X = DataFrame(data, columns=['Production_Budget_USD'])
Y = DataFrame(data, columns=['Worldwide_Gross_USD'])
```

```python
plt.figure(figsize=(11,7))#change the figure size to 10 x 6
plt.scatter(X,Y, alpha=0.3)#given a alpha value
#Given titles to the chart
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
#Changing 
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X,Y)
```

Slope Coefficient

```python
regression.coef_ #theta_1
```

```python
regression.intercept_ #intercept
```

```python
plt.figure(figsize=(11,7))#change the figure size to 10 x 6
plt.scatter(X,Y, alpha=0.3)#given a alpha value

plt.plot(X, regression.predict(X), color = 'red', linewidth = 4)

#Given titles to the chart
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
#Changing 
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

```python
#Goodness of Fit
regression.score(X,Y)
```
